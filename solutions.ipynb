```python
import pandas as pd  # keywords are highlighted in green, other strings in red, etc.
import numpy as np
```

```python
# Reading data into a DataFrame
file1 = pd.read_csv('./files_for_lab/csv_files/file1.csv')
#display(file1.head(10))
file2 = pd.read_csv('./files_for_lab/csv_files/file2.csv')
#display(file2.head(10))
file3 = pd.read_csv('./files_for_lab/csv_files/file3.csv')
display(file3.head(10))
```

```python
#Show the shape of the dataframes
display(file1.shape)
display(file2.shape)
display(file3.shape)
```

```python
#Standardize header names and ordering of columns.
file1_columns = file1.columns
file2_columns = file2.columns
file3_columns = file3.columns
display(file1_columns)
display(file2_columns)
display(file3_columns)
file2 = file2[['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
#renaming columns in file3
file3 = file3.rename(columns={'State':'ST'})
#order file3
file3 = file3[['Customer', 'ST', 'Gender', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
#lowercase file1 und file2 column GENDER
file1 = file1.rename(columns={'GENDER':'Gender'})
file2 = file2.rename(columns={'GENDER':'Gender'})
```

```python
#Deleting % sign in file1 and file2 before contatenating all three files
#file1['Customer Lifetime Value'] = file1['Customer Lifetime Value'].map(lambda x: x.lstrip('%'))
file1['Customer Lifetime Value'] = file1['Customer Lifetime Value'].str.replace("%", "") 
#Convert Customer Lifetime Value column to number
file1['Customer Lifetime Value'] =  pd.to_numeric(file1['Customer Lifetime Value'], errors='coerce')
file1
```

```python
#Checking datatypes file1
file1.dtypes
```

```python
#doing the same with file2
file2['Customer Lifetime Value'] = file2['Customer Lifetime Value'].str.replace("%", "")
#Convert Customer Lifetime Value column to number
file2['Customer Lifetime Value'] =  pd.to_numeric(file2['Customer Lifetime Value'], errors='coerce')
display(file2.dtypes)
```

```python
#Dividing file1 and file2 Customer Lifetime Value column by 100
file1['Customer Lifetime Value'] = file1['Customer Lifetime Value'] / 100
display(file1)
```

```python
file1['Customer Lifetime Value'] = file1['Customer Lifetime Value'] * 100 #had to do this because accidentally run cell above twice
file1
```

```python
#Dividing file2 Customer Lifetime Value column by 100
file2['Customer Lifetime Value'] = file2['Customer Lifetime Value'] / 100
display(file2)
```

```python
#concatenate all three files
all_data = pd.concat([file1,file2,file3], axis=0)
display(all_data)
```

```python
#Delete the column education and the number of open complaints from the dataframe.
all_data = all_data.drop(['Education','Number of Open Complaints'], axis=1)
display(all_data)
```

```python
#Reset index after concatenating.
all_data.reset_index()
display(all_data)
```

```python
#Check for duplicate rows in the data and remove if any.
all_data_copy = all_data.copy()
all_data_copy = all_data_copy.drop_duplicates()
print(all_data_copy.shape) #--> did this copying thing just for checking that it is working
display(all_data_copy)
all_data = all_data.drop_duplicates()
print(all_data.shape)
display(all_data)
```

```python
#Filter out the data for customers who have an income of 0 or less.
all_data = all_data[all_data['Income'] > 0]
display(all_data)
```
